```python
#Importation des fichiers csv
dispo_alim = pd.read_csv('DAN-P9-data/DisponibiliteAlimentaire_2017.csv', sep=',')
df_dispo_alim = pd.DataFrame(dispo_alim)

population = pd.read_csv('DAN-P9-data/Population_2000_2018.csv', sep=',')
df_population = pd.DataFrame(population)

macro_eco = pd.read_csv('Données_choisies/data_macro_eco.csv', sep=',')
df_macro_eco = pd.DataFrame(macro_eco)

stab_politique = pd.read_csv('Données_choisies/data_stab_politique.csv', sep=',')
df_stab_pol = pd.DataFrame(stab_politique)
```

```python
df_dispo_alim.head()
```

```python
df_dispo_alim = df_dispo_alim.pivot_table(
    index=['Zone', 'Année', 'Produit'],
    columns='Élément',
    values='Valeur',
    aggfunc='first'  # ou 'sum' si tu as plusieurs lignes pour une même combinaison
).reset_index()
```

```python
df_macro_eco['Année'] = df_macro_eco['Année'].replace('2016-2018', '2017')
df_macro_eco = df_macro_eco.pivot_table(
    index=['Zone', 'Année'],
    columns='Produit',
    values='Valeur',
    aggfunc='first'  # ou 'sum' si tu as plusieurs lignes pour une même combinaison
).reset_index()
```
